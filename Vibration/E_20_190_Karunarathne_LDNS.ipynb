**Derive the 2-dof freedom model that will capture the first two dominant modes of small amplitude vibration of the system.**

1. Continuous System Description

For small-amplitude transverse vibrations of a slender beam, the dynamics can be accurately described using the Euler--Bernoulli beam theory. The governing partial differential equation (PDE) of the continuous system is

\begin{equation}
\rho A \frac{\partial^2 y}{\partial t^2}
+ c \frac{\partial y}{\partial t}
+ EI \frac{\partial^4 y}{\partial x^4}
+ P \frac{\partial^2 y}{\partial x^2}
= f(t,x)
\end{equation}


where  
$y(x,t)$ is the transverse displacement,  
$\rho$ is the material density,  
$A$ is the cross-sectional area,  
$E$ is Young’s modulus,  
$I$ is the second moment of area,  
$c$ is the distributed viscous damping per unit length,  
$P$ is the axial compressive force, and  
$f(t,x)$ is the external transverse load per unit length.



2. Boundary conditions


The beam is assumed to have one end free and the base elastically restrained in rotation by a spring. The corresponding boundary conditions are

\begin{align}
y(0,t) &= 0, \qquad \frac{\partial^3 y}{\partial x^3}(l_{beam},t) = 0 \\
EI\frac{\partial^2 y}{\partial x^2}(0,t) - k_{\theta}\frac{\partial y}{\partial x}(0,t) &= 0,
\qquad \frac{\partial^2 y}{\partial x^2}(l_{beam},t) = 0
\end{align}

where the rotational stiffness is defined as

\begin{equation}
k_{\theta} = k_{spring} l_{spring}^2
\end{equation}




3. Operator Formulation

Let $\mathfrak{F}$ denote the infinite-dimensional vector space of four-times differentiable functions defined on the interval $[0,L]$. Define the linear operator $H:\mathfrak{F} \rightarrow \mathfrak{F}$ as

\begin{equation}
H = \frac{EI}{\rho A} \frac{\partial^4}{\partial x^4}
\end{equation}

Using this operator, the equations of motion can be written compactly as

\begin{equation}
\frac{\partial^2 y}{\partial t^2}
+ C \frac{\partial y}{\partial t}
+ H y
= q(t,x)
\end{equation}

where $c = C \rho A$ and $f(t,x) = \rho A q(t,x)$.

The operator $H$ maps a transverse deflection shape to the acceleration caused purely by beam bending, subject to the specified boundary conditions.

4. Free Vibration and Eigenvalue Problem

Assume that $H$ has a countably infinite set of distinct positive eigenvalues $\{\omega_k^2\}_{k=1}^{\infty}$ with corresponding eigenfunctions $\{\psi_k(x)\}_{k=1}^{\infty}$, representing the mode shapes of the beam.

The eigenvalue problem is therefore

\begin{equation}
H \psi_k(x) = \omega_k^2 \psi_k(x)
\end{equation}

Substituting the definition of $H$ yields

\begin{align}
\frac{EI}{\rho A} \frac{d^4 \psi_k(x)}{dx^4} &= \omega_k^2 \psi_k(x) \\
\frac{d^4 \psi_k(x)}{dx^4} &= \frac{\rho A \omega_k^2}{EI} \psi_k(x) \\
\frac{d^4 \psi_k(x)}{dx^4} &= \beta_k^4 \psi_k(x)
\end{align}

where

\begin{equation}
\beta_k^4 = \frac{\rho A \omega_k^2}{EI}
\end{equation}


5. General Solution of the Mode Shape Equation

The general solution of the fourth-order differential equation is

\begin{equation}
\psi_k(x) =
A_k \cosh(\beta_k x)
+ B_k \sinh(\beta_k x)
+ C_k \cos(\beta_k x)
+ D_k \sin(\beta_k x)
\end{equation}

where $A_k$, $B_k$, $C_k$, and $D_k$ are constants determined from the boundary conditions.

6. Application of Boundary Conditions

Base-End Conditions

From $\psi_k(0) = 0$,

\begin{equation}
A_k + C_k = 0
\end{equation}

The first and second derivatives at $x=0$ are

\begin{align}
\psi_k'(0) &= \beta_k (B_k + D_k) \\
\psi_k''(0) &= \beta_k^2 (A_k - C_k)
\end{align}

Using $C_k = -A_k$, the rotational boundary condition becomes

\begin{align}
EI \beta_k^2 (A_k - C_k) - k_{\theta} \beta_k (B_k + D_k) &= 0 \\
2 EI \beta_k^2 A_k - k_{\theta} \beta_k (B_k + D_k) &= 0
\end{align}

\subsubsection*{Free-End Conditions}

At $x = l_{beam}$, the boundary conditions yield

\begin{align}
\psi_k''(l_{beam}) = 0 \Rightarrow \quad
& A_k\left[\cosh(\beta_k l_{beam}) + \cos(\beta_k l_{beam})\right]
+ B_k \sinh(\beta_k l_{beam})
- D_k \sin(\beta_k l_{beam}) = 0 \\
\psi_k'''(l_{beam}) = 0 \Rightarrow \quad
& A_k\left[\sinh(\beta_k l_{beam}) - \sin(\beta_k l_{beam})\right]
+ B_k \cosh(\beta_k l_{beam})
- D_k \cos(\beta_k l_{beam}) = 0
\end{align}


7. Characteristic Equation

The above three equations form a homogeneous linear system

\begin{equation}
\mathbf{H}(\beta_k)
\begin{bmatrix}
A_k \\
B_k \\
D_k
\end{bmatrix}
= \mathbf{0}
\end{equation}

where $\mathbf{H}(\beta_k) \in \mathbb{M}_{3 \times 3}$.

For non-trivial solutions to exist, the determinant must vanish:

\begin{equation}
\det \mathbf{H}(\beta_k) = 0
\end{equation}

The roots $\beta_k$ are obtained numerically, and the corresponding constants $A_k$, $B_k$, and $D_k$ are found by solving the homogeneous system.

8. Reduced-Order (2-DOF) Approximation

The first two eigenfunctions $\psi_1(x)$ and $\psi_2(x)$ corresponding to the lowest eigenvalues $\omega_1^2$ and $\omega_2^2$ represent the first two dominant vibration modes. Retaining only these two modes provides a reduced-order 2-DOF model that captures the dominant dynamics of the beam.


Then, the eigenvalues of $H$ are given by,

$$
\{\omega_k^2\}_{k=1}^\infty = \left\{\frac{EI}{\rho A}{\beta_k}^4\right\}_{k=1}^\infty
$$

and the eigenfunctions/mode shapes of $H$ are given by,

$$
\{\psi_k\}_{k=1}^\infty = \left\{A_k \cosh(\beta_k x) + B_k \sinh(\beta_k x) + C_k \cos(\beta_k x) + D_k \sin(\beta_k x)\right\}_{k=1}^\infty
$$

$A_k$, $B_k$, $C_k$ and $D_k$ to be found.

###Forced Vibration Response Using Modal Expansion

1. External Forcing Model

Assuming there are no base excitations, the external transverse load acting on the beam is modelled as a harmonic point force applied at $x = l_{\text{exciter}}$:

\begin{equation}
q(x,t) = Q \sin(\Omega t)\,\delta(x - l_{\text{exciter}})
\end{equation}

2. Modal Decomposition of the Governing Equation

The governing equation of motion of the beam is

\begin{equation}
\ddot{y} + C\dot{y} + Hy = q(x,t)
\end{equation}

Using modal expansion, the system decouples into a set of forced, damped single-degree-of-freedom oscillators for each mode $k$:

\begin{equation}
m_k \ddot{z}_k
+ 2\zeta_k \omega_k m_k \dot{z}_k
+ m_k \omega_k^2 z_k
=
\langle\langle \psi_k(x), q(x,t) \rangle\rangle
\end{equation}

where

\begin{equation}
\zeta_k = \frac{C}{2\omega_k},
\qquad
m_k = \int_0^{l_{\text{beam}}} \rho A \, \psi_k^2(x)\,dx
\end{equation}

The modal mass $m_k$ normalises the arbitrary scaling introduced by the choice of mode shape constants.

3. Steady-State Modal Response

Solving the decoupled modal equation for harmonic excitation yields the steady-state response

\begin{equation}
z_k(t) = \chi_k(\Omega)\cos(\Omega t + \phi_k(\Omega))
\end{equation}

where $\chi_k(\Omega)$ is the steady-state amplitude and $\phi_k(\Omega)$ is the phase lag of the $k^{\text{th}}$ modal coordinate.



4. Evaluation of the Modal Forcing Term

The inner product defining the modal forcing is

\begin{align*}
\langle\langle \psi_k(x), q(x,t) \rangle\rangle
&= \int_0^{l_{\text{beam}}} \psi_k(x) q(x,t)\,dx \\
&= \int_0^{l_{\text{beam}}} \psi_k(x) Q \sin(\Omega t)\,\delta(x - l_{\text{exciter}})\,dx \\
&= Q \sin(\Omega t)\psi_k(l_{\text{exciter}})
\end{align*}


5. Steady-State Amplitude and Phase

The steady-state amplitude of the $k^{\text{th}}$ mode is given by

\begin{align*}
\chi_k(\Omega)
&=
\left|
\frac{\langle\langle \psi_k(x), q(x)\rangle\rangle}
{m_k \sqrt{(\omega_k^2 - \Omega^2)^2 + 4\zeta_k^2\Omega^2\omega_k^2}}
\right| \\
&=
\left|
\frac{Q \psi_k(l_{\text{exciter}})}
{m_k \sqrt{(\omega_k^2 - \Omega^2)^2 + 4\zeta_k^2\Omega^2\omega_k^2}}
\right|
\end{align*}

The phase lag is

\begin{equation}
\phi_k(\Omega)
=
\arctan\!\left(
\frac{2\zeta_k\Omega\omega_k}{\omega_k^2 - \Omega^2}
\right)
\end{equation}

6. Infinite-Dimensional Solution

The full displacement field of the beam is therefore expressed as

\begin{equation}
y(t,x)
=
\sum_{k=1}^{\infty} z_k(t)\psi_k(x)
=
\sum_{k=1}^{\infty}
\chi_k(\Omega)\cos(\Omega t + \phi_k(\Omega))\psi_k(x)
\end{equation}


###Reduction to a 2-DOF Model with Localised Spring--Damper Attachment

7. Two-Mode Approximation

Retaining only the first two dominant modes, the displacement is approximated by

\begin{align}
y(t,x) \approx\;
& \chi_1(\Omega)\cos(\Omega t + \phi_1(\Omega))\psi_1(x) \nonumber\\
&+ \chi_2(\Omega)\cos(\Omega t + \phi_2(\Omega))\psi_2(x)
\end{align}

8. Baseline 2-DOF Beam Model

The reduced-order system in modal coordinates is

\begin{align*}
\underbrace{
\begin{bmatrix}
m_1 & 0 \\
0 & m_2
\end{bmatrix}}_{M_{\text{beam}}}
\begin{bmatrix}
\ddot{z}_1 \\ \ddot{z}_2
\end{bmatrix}
+
\underbrace{
\begin{bmatrix}
2\zeta_1\omega_1 m_1 & 0 \\
0 & 2\zeta_2\omega_2 m_2
\end{bmatrix}}_{C_{\text{beam}}}
\begin{bmatrix}
\dot{z}_1 \\ \dot{z}_2
\end{bmatrix}
+
\underbrace{
\begin{bmatrix}
\omega_1^2 m_1 & 0 \\
0 & \omega_2^2 m_2
\end{bmatrix}}_{K_{\text{beam}}}
\begin{bmatrix}
z_1 \\ z_2
\end{bmatrix}
=
\underbrace{
\begin{bmatrix}
Q\psi_1(l_{\text{exciter}})\sin(\Omega t) \\
Q\psi_2(l_{\text{exciter}})\sin(\Omega t)
\end{bmatrix}}_{F_{\text{exciter}}}
\end{align*}

###Inclusion of Localised Attachments

9. Effective Mass of Spring Attachment

According to Rayleigh’s theory, one-third of the spring mass contributes to the system inertia. Including fixing mass,

\begin{equation}
m_{\text{eff}} = \frac{1}{3}m_{\text{spring}} + m_{\text{fixed}}
\end{equation}

The transverse acceleration at the attachment point is

\begin{align}
\ddot{y}(l_{\text{spring}},t)
=
\ddot{z}_1(t)\psi_1(l_{\text{spring}})
+
\ddot{z}_2(t)\psi_2(l_{\text{spring}})
\end{align}

The resulting inertia force is

\begin{equation}
F_{\text{inertia}}(t)
=
m_{\text{eff}}\ddot{y}(l_{\text{spring}},t)
\end{equation}

Projecting onto modal coordinates yields

\begin{align}
\begin{bmatrix}
F^{(m)}_1 \\
F^{(m)}_2
\end{bmatrix}
=
M_{\text{spring}}
\begin{bmatrix}
\ddot{z}_1 \\
\ddot{z}_2
\end{bmatrix}
\end{align}

with

\begin{align}
M_{\text{spring}}
=
m_{\text{eff}}
\begin{bmatrix}
\psi_1^2(l_{\text{spring}}) &
\psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) \\
\psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) &
\psi_2^2(l_{\text{spring}})
\end{bmatrix}
\end{align}

Additional mass matrices for the damper and exciter are similarly defined as

\begin{align}
M_{\text{damper}} &=
m_{\text{damper}}
\begin{bmatrix}
\psi_1^2(l_{\text{damper}}) &
\psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) \\
\psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) &
\psi_2^2(l_{\text{damper}})
\end{bmatrix} \\[2mm]
M_{\text{exciter}} &=
m_{\text{exciter}}
\begin{bmatrix}
\psi_1^2(l_{\text{exciter}}) &
\psi_1(l_{\text{exciter}})\psi_2(l_{\text{exciter}}) \\
\psi_1(l_{\text{exciter}})\psi_2(l_{\text{exciter}}) &
\psi_2^2(l_{\text{exciter}})
\end{bmatrix}
\end{align}


10. Damper Attachment Contribution

The viscous damper force is

\begin{equation}
F_{\text{damper}}(t)
=
c_{\text{damper}}\dot{y}(l_{\text{damper}},t)
\end{equation}

Using the two-mode approximation,

\begin{equation}
\dot{y}(l_{\text{damper}},t)
=
\dot{z}_1(t)\psi_1(l_{\text{damper}})
+
\dot{z}_2(t)\psi_2(l_{\text{damper}})
\end{equation}

The modal damping forces are

\begin{align}
F^{(c)}_k(t)
=
c_{\text{damper}}
\left(
\dot{z}_1\psi_1(l_{\text{damper}})
+
\dot{z}_2\psi_2(l_{\text{damper}})
\right)\psi_k(l_{\text{damper}})
\end{align}

which yields

\begin{align}
\begin{bmatrix}
F^{(c)}_1 \\
F^{(c)}_2
\end{bmatrix}
=
C_{\text{damper}}
\begin{bmatrix}
\dot{z}_1 \\
\dot{z}_2
\end{bmatrix}
\end{align}

with

\begin{align}
C_{\text{damper}}
=
c_{\text{damper}}
\begin{bmatrix}
\psi_1^2(l_{\text{damper}}) &
\psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) \\
\psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) &
\psi_2^2(l_{\text{damper}})
\end{bmatrix}
\end{align}

11. Final Reduced-Order System

Including all inertial and damping contributions, the final 2-DOF system is

\begin{equation}
(\mathbf{M}_{\text{beam}} + \mathbf{M}_{\text{spring}} + \mathbf{M}_{\text{damper}} + \mathbf{M}_{\text{exciter}})
\mathbf{\ddot{z}}
+
(\mathbf{C}_{\text{beam}} + \mathbf{C}_{\text{damper}})
\mathbf{\dot{z}}
+
\mathbf{K}_{\text{beam}}
\mathbf{z}
=
\mathbf{F}_{\text{exciter}}
\end{equation}

where $\mathbf{M}, \mathbf{C}, \mathbf{K} \in \mathbb{M}_{2\times2}$ and $\mathbf{z}, \mathbf{F} \in \mathbb{R}^2$.
